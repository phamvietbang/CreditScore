```python
def calculate_distribution(data, key):
    result = {
        "100":0,
        "200":0,
        "300":0,
        "400":0,
        "500":0,
        "600":0,
        "700":0,
        "800":0,
        "900":0,
        "1000":0
    }

    for i in data[key]:
        if i<100:
            result["100"]+=1
        elif i<200:
            result["200"]+=1
        elif i<300:
            result["300"]+=1
        elif i<400:
            result["400"]+=1
        elif i<500:
            result["500"]+=1
        elif i<600:
            result["600"]+=1
        elif i<700:
            result["700"]+=1
        elif i<800:
            result["800"]+=1
        elif i<900:
            result["900"]+=1
        else:
            result["1000"]+=1
    return result
```

```python
current_credit_score = calculate_distribution(data, "current_credit_score")
current_credit_score_con = calculate_distribution(data[data["number_of_liquidate"]>=3], "current_credit_score")
x = [i for i in current_credit_score.keys()]
y = [int(i) for i in current_credit_score.values()]
x1 = [i for i in current_credit_score_con.keys()]
y1 = [int(i) for i in current_credit_score_con.values()]
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].bar(x, y)
ax[1].bar(x1, y1)
sum_=sum(y)
sum_1=sum(y1)
# ax.set_yscale('log')
for i, v in enumerate(y):
    tmp=y[i]
    ax[0].text(i-.5,v+100,f"{tmp}\n{round(tmp*100/sum_, 2)}%",color='r')
for i, v in enumerate(y1):
    tmp=y1[i]
    ax[1].text(i-.5,v+10,f"{tmp}\n{round(tmp*100/sum_1, 2)}%",color='r')

ax[0].set_xlabel('Credit Score of All Wallets')
ax[1].set_xlabel('Credit Score of Wallets Liquidated More Than 3 Times')
ax[0].set_ylabel('Number of Wallets')
fig.suptitle('Current Credit Score Distribution')
plt.show()
```
